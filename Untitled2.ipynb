```python
df_loc = df_no_outliers[['employee_residence','company_location']]
df_loc.head()
```

```python
#creating an array to match values 

d =[]

for i,l in zip(df_loc["employee_residence"],df_loc["company_location"]):
    if i == l:
        d.append("Same")
    else:
        d.append("different")
```

```python
df_loc["Location"] = d
```

```python
df_loc.tail()
```

```python
df_loc = df_loc.drop(["employee_residence","company_location"], axis =1 )
```

```python
df_loc.head()
```

```python
#merging location back to df_no_outliers
df_no_outliers = pd.concat([df_no_outliers,df_loc],axis = 1)
```

```python
df_no_outliers.head()
```

```python
df_no_outliers = df_no_outliers.drop(["Same Location"], axis = 1)
```

```python
df_no_outliers.tail()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.barplot(data=df_no_outliers,x = "Location", y = "salary_in_usd",ax=ax).set(title='Salary based on Employee Location')
```
